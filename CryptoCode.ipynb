```python
import pandas as pd
import numpy as np

# reading the text file
#df = pd.read_csv('temp.csv')
df = pd.read_csv('CryptoData2.csv', sep = '\t')
df.head(15)
```

```python
# df.reset_index(drop=True).tail(5)
df['date'] = pd.to_datetime(df.date, infer_datetime_format = True)
display(df.head(15))
```

```python
df.sort_values(by = 'date', ascending = True, inplace = True)
display(df.head(15))
```

```python
# Convert the date to datetime64
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

traindata_df = df.loc[(df['date'] >= '2019-04-01') & (df['date'] <= '2019-10-31')]
traindata_df1 = df.loc[(df['date'] >= '2020-04-01') & (df['date'] <= '2020-10-31')]
traindata_df2 = df.loc[(df['date'] >= '2021-04-01') & (df['date'] <= '2021-10-31')]

testdata1_df = df.loc[(df['date'] >= '2019-01-01') & (df['date'] <= '2019-03-31')]
testdata1_df1 = df.loc[(df['date'] >= '2020-01-01') & (df['date'] <= '2020-03-31')]
testdata1_df2 = df.loc[(df['date'] >= '2021-01-01') & (df['date'] <= '2021-03-31')]

testdata2_df = df.loc[(df['date'] >= '2019-11-01') & (df['date'] <= '2019-12-31')]
testdata2_df1 = df.loc[(df['date'] >= '2020-11-01') & (df['date'] <= '2020-12-31')]
testdata2_df2 = df.loc[(df['date'] >= '2021-11-01') & (df['date'] <= '2021-12-31')]
  

```

```python
train = pd.concat([traindata_df,traindata_df1,traindata_df2],axis=0)
test1 = pd.concat([testdata1_df,testdata1_df1,testdata1_df2],axis=0)
test2 = pd.concat([testdata2_df,testdata2_df1,testdata2_df2],axis=0)
```
